```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.tsa.api import SimpleExpSmoothing
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt
```

```python
series=[2,5,0,2,4,1,2,2,5,7,2,5,2,3,4,3,3,5,8,2,5,3,5,4,6,4,7,3,8,3]
index=pd.date_range('2019-01-18', periods=30, freq='D')
data=pd.DataFrame(series, index=index)
data.columns=['Goals']
#Otimme tarkkailuun Chicago Blackhawks NHL-joukkueen maalimäärät sitten 18.1.2019
```

```python
fit1 = SimpleExpSmoothing(data['Goals']).fit()
data['Predict']=fit1.fittedvalues
data.plot()
#Maalien määrän trendi on nouseva
```

```python
sqrt(mean_squared_error(data['Goals'], data['Predict']))
#Ennustevirheiden keskihajonta on alhainen, mikä tarkoittaa että malli toimii.
```

```python
mean_absolute_error(data['Goals'], data['Predict'])
#Ennustevirheiden itseisarvojen keskiarvo
```

```python
data['Resid']=fit1.resid
data['Resid'].plot()
#Mallista näemmä ettei virheet noudata mitään tiettyä kaavaa
```

```python
index=pd.date_range('2019-02-17', periods=30, freq='D')
datap=pd.DataFrame(fit1.forecast(4), index=index)
datap.columns=['Predict']
data['Goals'].plot()
datap['Predict'].plot()
#Ennustimme Chicago Blackhawksien seuraavien pelien maalimääräksi 4-5 maalia. 
```
