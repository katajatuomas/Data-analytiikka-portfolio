```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
plt.style.use('seaborn-whitegrid')
```

```python
ttalo=pd.read_csv('ttalo1.csv', sep=';',decimal=',',skiprows=1)
ttalo.head()
```

```python
pihlis=pd.read_csv('pihlis1.csv', sep=';',decimal=',',skiprows=1)
pihlis.head()
```

```python
ttalo.index=pd.to_datetime(ttalo['Date'],dayfirst=True)
ttalo.head()
```

```python
pihlis.index=pd.to_datetime(ttalo['Date'],dayfirst=True)
pihlis.head()
```

```python
ttalo.sort_index(inplace=True)
ttalo.head()
```

```python
pihlis.sort_index(inplace=True)
pihlis.head()
```

```python
ttalo['Closing price'].plot()
```

```python
pihlis['Closing price'].plot()
```

```python
fig1, ax1=plt.subplots(figsize=(10,6))

color='C0'
ax1.set_ylabel('ttalo', color=color)
ax1.plot(ttalo['Closing price'],color=color)

ax2=ax1.twinx()
color='C1'
ax2.set_ylabel('pihlis', color=color)
ax2.plot(pihlis['Closing price'],color=color)
```

```python
ttalo['Closing price'].plot()
ttalo['Closing price'].rolling(100).mean().plot()
#Juoksevaa keskiarvoa voidaan kytt esimerkiksi ostosignaalin tunnistamiseen. 
#Terveystalon ostohetki tll metodilla olisi ollut esimerkiksi 2018 marraskuun ja 2019 tammikuun vliss, 
#kun osakekurssi "lvisti" juoksevan keskiarvon. Myyntihetki on taas kun kurssi laskee alle juoksevan keskiarvon.
```

```python
pihlis['Closing price'].plot()
pihlis['Closing price'].rolling(100).mean().plot()
```

```python
ttalo['TTALO_change']=ttalo['Closing price'].pct_change()
ttalo.head()
```

```python
pihlis['PIHLIS_change']=pihlis['Closing price'].pct_change()
pihlis.head()
```

```python
muutokset=pd.concat([ttalo['TTALO_change'],pihlis['PIHLIS_change']],axis=1)
muutokset.head()
```

```python
ax3=muutokset['2018-01':].plot()
ax3.set_ylabel('Muutos')
vals=ax3.get_yticks()
ax3.set_yticklabels(['{:.0f} %'.format(y*100) for y in vals])
```

```python
muutokset.describe()
```

```python
muutokset[(muutokset ['TTALO_change']<-0.05)|(muutokset['PIHLIS_change']>0.05)]
```

```python
muutokset.corr()
#Korrelaatiokerroin osakekurssien vlill on jopa hmmstyttvn pieni. Yritykset ovat kuitenkin kilpailijoita samalla alalla
```

```python
korrelaatio=muutokset['TTALO_change'].rolling(200,min_periods=30).corr(muutokset['PIHLIS_change'])
korrelaatio.plot()
#Korrelaatiokerroin oli 2018 alkupuolella korkeampi, mutta syyskuussa 2019 tapahtui merkittv muutos. 
```
